```python
# lets make a function for generating the feedback content for the dataset

def generate_feedback_content(prompt,scam=True):
    completion = client.chat.completions.create(
        model="llama3-8b-8192",
        messages=[
            {
                "role": "system",
                "content": "You are a cybersecurity expert. " +
                "You are helping generate a dataset for a training program. "+
                "You are generating the feedback content that will be used to annotate the dataset."+
                "You don't know if the user got it right or wrong, so don't congratulate on a correct answer."+
                "Just say whether it is a scam or not (based on what you are prompted it is or isn't), and then explain why."

            },
            {
                "role": "user",
                "content": 
                "This mesage is a scam." if scam else "This message is not a scam."+
                "Please explain why."+
                "Also, identify the type of scam" if scam else ""+
                f"\n{prompt}"
            }
        ],
        temperature=1,
        max_tokens=1024,
        top_p=1,
        stream=False,
        stop=None,
    )

    return completion.choices[0].message.content.replace('\n', '\\n')
```

```python
import pandas as pd
from tqdm import tqdm

def process_emails(input_csv, output_csv, scam=True):
    # read the input CSV file
    real_emails = pd.read_csv(input_csv, header=None)

    # set the column name of the first column to "email content"
    real_emails.columns = ['email_content']

    # create a new column for the feedback content
    tqdm.pandas(desc="Generating feedback content")
    real_emails['feedback_content'] = real_emails['email_content'].progress_apply(lambda x: generate_feedback_content(x, scam))

    # save the new dataframe to the output CSV file
    real_emails.to_csv(output_csv, index=False)

# Example usage
# process_emails('real_emails.csv', 'real_emails_with_feedback.csv', scam=False)
```

```python
process_emails('real_emails.csv', 'real_emails_with_feedback.csv', scam=False)
process_emails('scam_emails.csv', 'scam_emails_with_feedback.csv', scam=True)
process_emails('real_texts.csv', 'real_texts_with_feedback.csv', scam=False)
process_emails('scam_texts.csv', 'scam_texts_with_feedback.csv', scam=True)
```

```python
# go through the items in the combined_data.json
# for each item that is has the category "text": randomly generate a phone number as an attribute labelled "senderNumber"
# for each item that is has the category "email": use regex to extract the first email address included in the "message" atrtibute as an attribute labelled "senderEmail"

import re
import random

with open('combined_data.json', 'r') as file:
    data = json.load(file)

for item in data:
    if item["type"] == "text":
        item["senderNumber"] = f"{random.randint(100, 999)}-{random.randint(100, 999)}-{random.randint(1000, 9999)}"
    elif item["type"] == "email":
        email = re.search(r'[\w\.-]+@[\w\.-]+', item["message"])
        if email:
            item["senderEmail"] = email.group()


# save the updated data to a new file
with open('combined_data_updated.json', 'w') as file:
    json.dump(data, file, indent=4)
```
