```python
# read in real_emails_with_feedback.csv

import csv
import os
import sys
import re
import datetime
import time

# read in the csv file
def read_csv(file):
    with open(file
                , 'r'
                , encoding='utf-8'
                , errors='ignore') as f:
            reader = csv.reader(f)
            data = list(reader)


    return data


# print first 5 rows
def print_first_5_rows(data):
    for i in range(5):
        print(data[i])
```

```python
import pandas as pd

def delete_scam_emails(data):
    # Filter out rows where the "scam" column is "1"
    filtered_data = data[data['scam'] != 1]
    return filtered_data

# Read the CSV file
data = pd.read_csv('real_emails_with_feedback.csv')

# Apply the function to delete scam emails
filtered_data = delete_scam_emails(data)

# Print the first 5 rows of the filtered data
print(filtered_data.head())

# save the filtered data to a new CSV file
filtered_data.to_csv('scam_emails_feedback_edited', index=False)
```

```python
# can you write a function that converts a csv file into a json file such that the output is in the correct JSON format (i.e. with the correct nesting of keys and values)?
import json

def csv_to_json(csv_file, json_file):
    data = read_csv(csv_file)
    keys = data[0]
    values = data[1:]
    json_data = []
    for value in values:
        json_data.append(dict(zip(keys, value)))

    with open(json_file, 'w') as f:
        json.dump(json_data, f, indent=4)

csv_to_json('real_emails_feedback_edited.csv', 'real_emails_feedback.json')

```

```python
csv_to_json('scam_emails_with_feedback.csv', 'scam_emails_feedback.json')
csv_to_json('real_texts_with_feedback.csv', 'real_texts_feedback.json')
csv_to_json('scam_texts_with_feedback.csv', 'scam_texts_feedback.json')
```

```python
# combine all of the json files into one
def combine_json_files(files, output_file):
    combined_data = []
    for file in files:
        with open(file, 'r') as f:
            data = json.load(f)
            combined_data.extend(data)

    with open(output_file, 'w') as f:
        json.dump(combined_data, f, indent=4)

combine_json_files(['real_emails_feedback.json', 'scam_emails_feedback.json', 'real_texts_feedback.json', 'scam_texts_feedback.json'], 'combined_data.json')
```
